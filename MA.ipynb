```python

b = 7
start = b - 1
price = df.Price.tolist()
date = df.Date.tolist()
Price = np.asarray(price)
Date = np.asarray(date)
sma1 = talib.SMA(Price, timeperiod = b)
Trades = []
Dates =[]
Exit = []
for i in range(start, len(price)):
    i = start
    if i+1 == len(price):
        break
    else:
        if price[i] > sma1[i]:
            Trades.append(price[i])
            Dates.append(date[i])
            next = i + 1
            for i in range(next, len(price)):
                start = i
                if price[i] < sma1[i]:
                    Trades.append(price[i])
                    Dates.append(date[i])
                    Trades.append('B')
                    start = i
                    break
        if price[i] < sma1[i]:
            Trades.append(price[i])
            Dates.append(date[i])
            next = i + 1
            for i in range(next, len(price)):
                start = i
                if price[i] > sma1[i]:
                    Trades.append(price[i])
                    Dates.append(date[i])
                    Trades.append('S')
                    break
if len(Trades) % 3 ==1:
    Trades.pop(-1)
if len(Dates) % 2 ==1:
    Dates.pop(-1)
Entry = []
for i in range(0,len(Trades),3):
    Entry.append(Trades[i])
Exit = []
for i in range(1,len(Trades),3):
    Exit.append(Trades[i])
Signal = []
for i in range(2,len(Trades),3):
    Signal.append(Trades[i])
EntryDate = []
ExitDate   = []
for i in range(0,len(Dates),2):
    EntryDate.append(Dates[i])
for i in range(1,len(Dates),2):
    ExitDate.append(Dates[i])
PnL = []   
for i in range(len(Signal)):
    if Signal[i] == 'B':
        value = Exit[i] - Entry[i]
        PnL.append(value)
        
    else: 
        value = Entry[i] - Exit[i]
        PnL.append(value)
MA = pd.DataFrame(
    {'EntryDate': EntryDate,
     'Entry': Entry,
     'ExitDate': ExitDate,
     'Exit': Exit,
     'Signal': Signal,
     'PnL': PnL
    })
        
        
        
```
